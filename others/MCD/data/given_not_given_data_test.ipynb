```python
import os
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams["figure.figsize"] = (10,12)
```

```python
given_data = np.load('./given.npy',allow_pickle=True)
not_given_data = np.load('./not_given.npy',allow_pickle=True)
optimal_cost = 10.0
```

```python
given_total = len(given_data)
given_success = 0
given_fail = 0
for i in given_data:
    if i[4]:
        given_success+=1
    else:
        given_fail+=1
print("given success rate(percentage): ", int(given_success/given_total*100),'%')
print("given fail rate(percentage): ", int(given_fail/given_total*100),'%')
```

```python
not_given_total = len(not_given_data)
not_given_success = 0
not_given_fail = 0
for i in not_given_data:
    if i[4]:
        not_given_success+=1
    else:
        not_given_fail+=1
print("not_given success rate(percentage): ", int(not_given_success/not_given_total*100),'%')
print("not_given fail rate(percentage): ", int(not_given_fail/not_given_total*100),'%')
```

```python
given_cost = []
for i in given_data:
    if i[4]:
        given_cost.append(i[3])
given_cost = np.array(given_cost)
print("given mean cost value: ",given_cost.mean())
print("given min cost value: ",given_cost.min())
print("given max cost value: ",given_cost.max())
print("given variance cost value: ",given_cost.var())
```

```python
not_given_cost = []
for i in not_given_data:
    if i[4]:
        not_given_cost.append(i[3])
not_given_cost = np.array(not_given_cost)
print("not_given mean cost value: ",not_given_cost.mean())
print("not_given min cost value: ",not_given_cost.min())
print("not_given max cost value: ",not_given_cost.max())
print("not_given variance cost value: ",not_given_cost.var())
```

```python
mins = np.array([given_cost.min(),not_given_cost.min()])
maxes = np.array([given_cost.max(),not_given_cost.max()])
means = np.array([given_cost.mean(),not_given_cost.mean()])
std = np.array([given_cost.var(),not_given_cost.var()])

# create stacked errorbars:
plt.errorbar(['given', 'not_given'], means, std, fmt='ok', lw=3)
plt.errorbar(['given', 'not_given'], means, [means - mins, maxes - means],fmt='.k', ecolor='gray', lw=1)
plt.show()
```
